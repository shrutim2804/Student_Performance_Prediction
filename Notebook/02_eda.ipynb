```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('../Dataset/cleaned_data.csv')
print("Shape:", df.shape)
df.head()
```

```python
num_cols = ['Age', 'StudyTimeWeekly', 'Absences', 'GPA']

plt.figure(figsize=(14, 4))
for i, col in enumerate(num_cols):
    plt.subplot(1, 4, i+1)
    sns.histplot(df[col], kde=True, color='steelblue')
    plt.title(col)
plt.tight_layout()
plt.savefig('../Documentation/histograms.png')
plt.show()
print("✅ Histograms saved")
```

```python
plt.figure(figsize=(7, 4))
grade_map = {0.0: 'A', 1.0: 'B', 2.0: 'C', 3.0: 'D', 4.0: 'F'}
df['Grade'] = df['GradeClass'].map(grade_map)
sns.countplot(x='Grade', data=df, palette='Set2', order=['A','B','C','D','F'])
plt.title('Grade Distribution (Target Variable)')
plt.xlabel('Grade')
plt.ylabel('Count')
plt.savefig('../Documentation/class_distribution.png')
plt.show()
df.drop(columns=['Grade'], inplace=True)
print("✅ Class distribution saved")
```

```python
plt.figure(figsize=(14, 4))
for i, col in enumerate(num_cols):
    plt.subplot(1, 4, i+1)
    sns.boxplot(x='GradeClass', y=col, data=df, palette='Set3')
    plt.title(f'{col} vs Grade')
plt.tight_layout()
plt.savefig('../Documentation/bivariate_boxplots.png')
plt.show()
print("✅ Bivariate boxplots saved")
```

```python
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.scatterplot(x='StudyTimeWeekly', y='GPA', hue='GradeClass', data=df, palette='Set1')
plt.title('Study Time vs GPA')

plt.subplot(1, 2, 2)
sns.scatterplot(x='Absences', y='GPA', hue='GradeClass', data=df, palette='Set1')
plt.title('Absences vs GPA')

plt.tight_layout()
plt.savefig('../Documentation/scatter_plots.png')
plt.show()
print("✅ Scatter plots saved")
```

```python
plt.figure(figsize=(12, 8))
corr = df.corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.savefig('../Documentation/heatmap.png')
plt.show()
print("✅ Heatmap saved")
```

```python
cat_cols = ['Tutoring', 'Extracurricular', 'Sports', 'Music', 'Volunteering']

plt.figure(figsize=(16, 4))
for i, col in enumerate(cat_cols):
    plt.subplot(1, 5, i+1)
    sns.countplot(x=col, hue='GradeClass', data=df, palette='Set2')
    plt.title(col)
    plt.legend([], frameon=False)
plt.tight_layout()
plt.savefig('../Documentation/categorical_analysis.png')
plt.show()
print("✅ Categorical analysis saved")
```
